# 3. Data Cleaning

## Code Example
In this section, we will first explore our data by checking for missing values and duplicated data. We will then employ some common techniques of dealing with them. We will use the **Pandas** and **Numpy** libraries.

```python
# check duplicated records
df[df.duplicated()]
```

We see that 3 records have been marked as duplicated. Since there is no benefit to having duplicated records, we will simply drop them.

```python
# drop duplicates
df.drop_duplicates(inplace=True)

# check shape of dataset
df.shape
```

```python
# check number of missing values in all columns
df.isna().sum()
```

```python
# check if there are any patterns in the missing values
df[df.isnull().any(axis=1)]
```

There are 3 main ways you can handle missing values.
* dropping the records with the missing values
* replacing the missing values with the column mean/median (for numeric data) or mode (for categorical data)
* replacing the missing values with specific values, e.g., 0 or "N"

Which method you choose will depend on your requirements and the data itself. In our case, we will carry out the following:
* Replace missing values for `Study_Hours` with the column mean
* Replace missing values for `Test_Prep_Hours` with the column median
* Replace missing values for `Prev_Final_Grade` with the column mode
* Replace missing values for `Internet_Access` and `Extracurriculars` with "N"

```python
# replace missing Study_Hours values with mean
df['Study_Hours'] = df['Study_Hours'].fillna(df['Study_Hours'].mean())

# replace missing Test_Prep_Hours values with median
df['Test_Prep_Hours'] = df['Test_Prep_Hours'].fillna(df['Test_Prep_Hours'].median())

# replace Prev_Final_Grade with mode
df['Prev_Final_Grade'] = df['Prev_Final_Grade'].fillna(df['Prev_Final_Grade'].mode().iloc[0])

# replace missing Internet_Access and Extracurriculars values with "N"
df['Internet_Access'] = df['Internet_Access'].fillna("N")
df['Extracurriculars'] = df['Extracurriculars'].fillna("N")

# confirm all missing values have been taken care of
df.isna().sum()
```

# 5. Data Modeling

## Code Example
There are various modeling algorithms in data science and machine learning. We will look at one of the most popular ones: **classification**. Classification is where we train a model to analyze the features of a record and place it in a category, class, or group.

Remember that our dataset has some categorical variables. Machine learning models don't work with non-numerical data so we need to convert the categorical values into discrete numerical values. This process is called **encoding** and there are various ways to perform this:
* **Manually:** you can manually replace the values using `np.where`. While this approach is the most straightforward, it is not feasible when you have many categorical variables that need to be encoded.
* **Label encoding:** you can use Scikit-learn's `LabelEncoder` which converts the individual categorical values into discrete numbers starting with 0. Take note that label encoding assumes there is an order or ranking system among the variables, and will assign weights accordingly. For example, "Low" will be given a value of 0, "Medium" will be given a value of 1, and "High" will be given a value of 2. This can affect the modelling results so use it only when there is a clear order in the variables.
* **One-hot encoding:** this is best used with variables that don't have an order or rank, and do not have too many categories. Keep in mind that this method will increase the dimensionality of your dataset (make it bigger by adding more columns).

In our example, we will apply manual encoding to `Gender`, one-hot encoding to `Internet_Access` and `Extracurriculars`, and label encoding to `Prev_Midterm_Grade`, `Prev_Final_Grade`, `Midterm_Grade`, and `Final_Grade`. We will also drop the `StudentID` column since it does not really benefit us and we will remove any new duplicated records that might be created as a result. 

```python
# manually encode Gender column
df["Gender"] = df["Gender"].map({"F": 1, "M": 0})

# confirm the Gender column has been encoded
df['Gender'].value_counts()
```

```python
# import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

# create an instance of OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# fit and transform the 'Internet_Access' column
df['Internet_Access'] = encoder.fit_transform(df[['Internet_Access']])

# fit and transform the 'Extracurriculars' column
df['Extracurriculars'] = encoder.fit_transform(df[['Extracurriculars']])

# preview the encoded variables
df[['Internet_Access', 'Extracurriculars']].head()
```

```python
# import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# create an instance of LabelEncoder
encoder = LabelEncoder()

# apply label encoding for 'Prev_Midterm_Grade'
df['Prev_Midterm_Grade'] = encoder.fit_transform(df['Prev_Midterm_Grade'])

# apply label encoding for 'Prev_Final_Grade'
df['Prev_Final_Grade'] = encoder.fit_transform(df['Prev_Final_Grade'])

# apply label encoding for 'Midterm_Grade'
df['Midterm_Grade'] = encoder.fit_transform(df['Midterm_Grade'])

# apply label encoding for 'Final_Grade'
df['Final_Grade'] = encoder.fit_transform(df['Final_Grade'])

# preview the encoded variables
df[['Prev_Midterm_Grade', 'Prev_Final_Grade', 'Midterm_Grade', 'Final_Grade']].head()
```

```python
# drop StudentID column
df.drop('StudentID', axis=1, inplace=True)

# drop duplicates
df.drop_duplicates(inplace=True)

# check dataset shape
print("New dataset shape:", df.shape)

# preview datatypes
df.dtypes
```

Our dataset is now ready for modeling.

We will split our dataset into two: the training set (which is used to train the model) and the testing set (which is used to test the model to see whether it has learned well enough to be able to generalize onto unseen data). By convention, we split them 80-20, that is, 80% of the data becomes the training set and the remaining 20% becomes the test set, but this can change depending on your requirements.

We will use Scikit-learn's `train_test_split` module for this.

```python
# import module
from sklearn.model_selection import train_test_split

# extract the independent variables (X) and the target/predictor variable (y)
X = df.drop(columns=['Final_Grade'], axis=1)
y = df['Final_Grade']

# split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Now that we have split our dataset into training and test sets, we will apply a classification algorithm. There are many classification algorithms in machine learning such as:
* Logistic regression
* Decision trees
* Random forest
* k-nearest neighbors
* Naive Bayes

For our example, we will use logistic regression.

```python
# import Logistic Regression
from sklearn.linear_model import LogisticRegression

# create an instance of LogisticRegression
logreg = LogisticRegression()

# fit the logistic regression model
logreg.fit(X_train, y_train)

# predict on the test set
y_pred = logreg.predict(X_test)
```

After we have carried out modeling, we can analyze how the model performed using a confusion matrix and the accuracy score, both of which are part of Sci-kit. 

In a confusion matrix for a binary outcome (e.g. True/False, Yes/No, etc.), the confusion matrix is typically organized as a matrix of four values, two rows and two columns:
* The first value (first row, first column) is the **True Negative (TN)** which represents the number of correctly predicted negative instances.
* The second value (first row, second column) is the **False Positive (FP)** which represents the number of incorrectly predicted positive instances.
* The third value (second row, first column) is the **False Negative (FN)** which represents the number of incorrectly predicted negative instances.
* The last value (second row, second column) is the **True Positive (TN)** which represents the number of correctly predicted positive instances.

```python
# import libraries
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# create confusion matrix
cm = confusion_matrix(y_test, y_pred)

# create a data frame from the confusion matrix
df_cm = pd.DataFrame(cm, index=['Actual 0', 'Actual 1', 'Actual 2', 'Actual 3'], columns=['Predicted 0', 'Predicted 1', 'Predicted 2', 'Predicted 3'])
df_cm
```

Here we see that accuracy is approximately 66.67%.

The confusion matrix you provided has a dimension of 4x4. Each row in the matrix represents the actual classes, and each column represents the predicted classes. Here's how you can interpret the confusion matrix:

- The first row indicates the instances of the actual class 0. In this case, there are no instances of class 0 in the predictions, as all the values in the row are 0.
- The second row represents the instances of the actual class 1. According to the matrix, there is 1 instance of class 1 that is correctly predicted (in the diagonal), while 1 instance of class 1 is misclassified as class 2, 2 instances are misclassified as class 3, and 0 instances are misclassified as class 0.
- The third row corresponds to the instances of the actual class 2. All the instances of class 2 (6 instances) are correctly predicted as class 2, as there are no misclassifications in this row (all values except the diagonal are 0).
- The fourth row represents the instances of the actual class 3. One instance of class 3 is correctly predicted as class 3, while one instance is misclassified as class 2. The other columns in this row have 0 values, indicating no misclassifications for other classes.

Overall, the confusion matrix shows the performance of a multi-class classification model on a specific dataset, where the diagonal elements represent correct predictions, and the off-diagonal elements represent misclassifications.
